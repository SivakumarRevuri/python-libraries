```python
people  = {
    'first' : ['Siva', "Moksha", "Lucky", "Pavan"],
    'last' : ['kumar', 'Sree', 'Papa', 'kalyan'],
    'email': ['sivakur@gmail.com', 'moksh2003@gmail.com', 'lucky@gmail.com', 'pkar@gmail.com']
}
```

```python
import pandas as pd

df = pd.DataFrame(people)
```

```python
df
```

```python
# By default, it shows numbers as indexes, we can customize that by using
# set_index() option
df.set_index('email')
```

```python
df
```

```python
# for parmenent change of index, we have inplace parameter as True.
df.set_index('email', inplace=True)
```

```python
df # see, now email will be considered as default index
```

```python
df.sort_index(inplace=True) # sorting the index parmenently
```

```python
df
```

```python
# and, we can reset the index to default by using df.rest_index() method
df.reset_index() # temporary
```

```python
df
```

```python
#for parmanenet
df.reset_index(inplace=True)
```

```python
df
```
